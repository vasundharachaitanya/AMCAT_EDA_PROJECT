```python
Q1 = df['10percentage'].quantile(0.25)
Q3 = df['10percentage'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out values beyond the bounds
df = df[(df['10percentage'] > lower_bound) & (df['10percentage'] < upper_bound)]
df['10percentage'].plot(kind = 'box')
```

**Observation:**
- The lowest percentage among employees is 43, while the highest percentage is 97.76.
- The average percentage of employees in the 10th percentile is 77.92.
- Since the employees with percentages less than 50 do not significantly affect the employees' salaries, outliers are handled using the IQR method.

### Problem Statement : 
#### “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.”

```python
# Assuming 'Role' column contains job roles and 'Salary' column contains salary information
computer_science_roles = ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer']

# Step 1: Filter Data
filtered_data = df[df['Designation'].isin(computer_science_roles)]

# Step 2: Calculate Salary Statistics
salary_statistics = filtered_data['Salary'].describe()

# Step 3: Compare with Claim
lower_limit = 250000  # Lower limit of the claimed salary range
upper_limit = 300000  # Upper limit of the claimed salary range

# Check if the calculated mean or median falls within the claimed range
claim_verified = (salary_statistics['mean'] >= lower_limit) and (salary_statistics['mean'] <= upper_limit)

# Display results
print("Claim Verification Results:")
print(f"Mean Salary: {salary_statistics['mean']}")
print(f"Median Salary: {salary_statistics['50%']}")
print(f"Claim Verified: {claim_verified}")
```

```python
import numpy as np
from scipy.stats import ttest_1samp

# Collect data (replace this with your actual data)
data = np.array([250000, 270000, 280000, 260000, 290000])

# Claimed mean from the statement
claimed_mean = 250000

# Set the significance level
alpha = 0.05

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, claimed_mean)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to suggest that the average salary is greater than the claimed value.")
else:
    print('Fail to reject the null hypothes. There is No enough evidence to suggest that the average salary is greater than the claimed value.')
```
